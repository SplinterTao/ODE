```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def softmax(x):
    exponentials=np.exp(x)
    standardize=np.sum(exponentials,axis=1)
    shape=x.shape
    for i in range(shape[0]):
        exponentials[i,]=exponentials[i,]/standardize[i]
    return exponentials
```

```python
def findmax(x):
        x_copy=np.zeros(shape=x.shape[0])
        for i in range(x.shape[0]):
            for j in range(x.shape[1]):
                if x[i,j]==max(x[i,]):
                    x_copy[i]=j
        return x_copy

from sklearn.model_selection import train_test_split
```
